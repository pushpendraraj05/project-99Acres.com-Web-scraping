```python
bath = df['bhk'].tolist()

print(bath)
```

```python
import re

# Define the pattern to match
pattern = '\d+'

# Use the finditer() function to check each item in the list against the pattern
numerical_values = [re.finditer(pattern, item) for item in bath]

#iterate through the second element of the iterator
numerical_values = [next(val).group() if val else None for val in numerical_values]

print(numerical_values)
```

```python
B_column = df['location'].tolist()

print(B_column)
```

```python
import re
```

```python
pattern = '\d+'
bhk = list(map(lambda x: re.search(pattern, x).group() if re.search(pattern, x) else None, B_column))
pattern = re.compile('\d+')
bhk = list(map(lambda x: pattern.search(x).group() if pattern.search(x) else None, B_column))
bhk
```

```python
pattern = '\d+'
numerical_values = list(map(lambda x: re.findall(pattern, x), B_column))
pattern = re.compile('\d+')
numerical_values = list(map(pattern.findall, B_column))
numerical_values
```
